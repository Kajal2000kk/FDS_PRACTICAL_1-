# Box and whiskers plot
Box and Whiskers plot for numerical vs categorical variable

A Box plot is also known as whisker plot created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum. In the box plot, a box is created from the first quartile to the third quartile, a vertical line is also there which goes through the box at the median. here x-axis denotes the data to be plotted while the y-axis shows the frequency distribution.

```python
sns.boxplot(x=cars_data['FuelType'],y=cars_data['Price'])
```

```python
sns.boxplot(x="FuelType",y=cars_data["Price"], hue="Automatic", data=cars_data)
```

```python
f,(ax_box,ax_hist)=plt.subplots(2,gridspec_kw={"height_ratios":(.15,.85)})
```
